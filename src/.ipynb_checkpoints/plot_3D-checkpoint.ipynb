```python
import csv
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
```

```python
reader = csv.reader(open("../data/data_200_G1.csv", "rb"), delimiter=",")
x = np.array(list(reader))
print x.shape

#remove header and select only 3 emotions data
x = x[1:len(x),1:4] 

#remove empty row -> which means the difference of rows of x and data is no. of clips
data = np.array([xx for xx in x if not '' in xx])

#cast type and divide by 100
excitement = []
for d in data:
    r = []
    for e in d:
        r.append(float(e) / 100)
    excitement.append(r)
excitement = np.array(excitement)

print excitement.shape, excitement
```

```python
reader = csv.reader(open("../data/data_200_G2.csv", "rb"), delimiter=",")
x = np.array(list(reader))
print x.shape

#remove header and select only 3 emotions data
x = x[1:len(x),1:4] 

#remove empty row -> which means the difference of rows of x and data is no. of clips
data = np.array([xx for xx in x if not '' in xx])

#cast type and divide by 100
happiness = []
for d in data:
    r = []
    for e in d:
        r.append(float(e) / 100)
    happiness.append(r)
happiness = np.array(happiness)

print happiness.shape, happiness
```

```python
reader = csv.reader(open("../data/data_200_G3.csv", "rb"), delimiter=",")
x = np.array(list(reader))
print x.shape

#remove header and select only 3 emotions data
x = x[1:len(x),1:4] 

#remove empty row -> which means the difference of rows of x and data is no. of clips
data = np.array([xx for xx in x if not '' in xx])

#cast type and divide by 100
fear = []
for d in data:
    r = []
    for e in d:
        r.append(float(e) / 100)
    fear.append(r)
fear = np.array(fear)

print fear.shape, fear
```

```python
# x_axis = Happiness value
column_id = 0
x_axis = np.append(np.append(excitement[:,column_id], happiness[:,column_id]), fear[:,column_id]) 
print x_axis.shape

# y_axis = Fear value
column_id = 1
y_axis = np.append(np.append(excitement[:,column_id], happiness[:,column_id]), fear[:,column_id]) # Fear
print y_axis.shape

# z_axis = Excitement value
column_id = 2
z_axis = np.append(np.append(excitement[:,column_id], happiness[:,column_id]), fear[:,column_id]) # Excitement
print y_axis.shape

colors = ['r'] * len(excitement) + ['b'] * len(happiness) + ['g'] * len(fear)
print len(colors)

markers = ['o'] * len(excitement) + ['^'] * len(happiness) + ['s'] * len(fear)
print len(markers)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x_axis, y_axis, z_axis, c=colors)

ax.set_xlabel('Happiness')
ax.set_ylabel('Fear')
ax.set_zlabel('Excitement')

# ax.legend(('Happiness', 'Fear', 'Excitement'), shadow=True, fancybox=True)
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

happy_id = 0
fear_id = 1
excitement_id = 2

ax.scatter(happiness[:,happy_id], happiness[:,fear_id], happiness[:,excitement_id], c='r', label = 'happiness')
ax.scatter(fear[:,happy_id], fear[:,fear_id], fear[:,excitement_id], c='b', label = 'fear')
ax.scatter(excitement[:,happy_id], excitement[:,fear_id], excitement[:,excitement_id], c='g', label = 'excitement')

ax.set_xlabel('Happiness')
ax.set_ylabel('Fear')
ax.set_zlabel('Excitement')

ax.legend(('Happiness', 'Fear', 'Excitement'), shadow=True, fancybox=True, loc = 'upper right')

for x in range(0, 360, 30):
    for y in range(0, 360, 30):
        ax.view_init(x, y)
#         plt.savefig('../figure3d/fig_' + str(x) + '_' + str(y) + '.png')
        plt.show()
#     break
```
