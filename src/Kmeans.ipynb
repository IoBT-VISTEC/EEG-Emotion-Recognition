```python
import numpy as np
import pandas as pd
from sklearn.metrics import silhouette_samples, silhouette_score, davies_bouldin_score
from sklearn.mixture import GaussianMixture
from sklearn.cluster import KMeans
```

```python
Dat = np.load('../data/score/raw.npy')
Dat = Dat.reshape([Dat.shape[0]*Dat.shape[1],Dat.shape[2]])
DatA = Dat[:,3]
DatV = Dat[:,4]
DatR = Dat[:,5]

X = DatA
Y = DatV
Z = DatR

X = X.reshape([-1,1])
Y = Y.reshape([-1,1])
Z = Z.reshape([-1,1])

XY = np.concatenate((Y,X),axis=1)
XYZ = np.concatenate((X,Y,Z),axis=1)
```

```python
kmeans = KMeans(n_clusters=6)
kmeans = kmeans.fit(XY)
labels = kmeans.predict(XY)
centroids = kmeans.cluster_centers_
```

```python
sse = []
list_k = list(range(1, 10))

for k in list_k:
    km = KMeans(n_clusters=k)
    km.fit(XY)
    sse.append(km.inertia_)

# Plot sse against k
plt.figure(figsize=(6, 6))
plt.plot(list_k, sse, '-o')
plt.title('The Elbow Method showing the optimal K')
plt.xlabel(r'Number of clusters K')
plt.ylabel('Sum of squared distance');
```

```python
from sklearn.metrics import silhouette_samples, silhouette_score
for i, k in enumerate([2,3,4,5,6]):
    
    fig, (ax1) = plt.subplots(1)
    #fig.add_subplot(111, projection='3d')
    #fig.set_size_inches(18, 7)
    
    # Run the Kmeans algorithm
    km = KMeans(n_clusters=k)
    labels = km.fit_predict(XY)
    centroids = km.cluster_centers_

    # Get silhouette samples
    silhouette_vals = silhouette_samples(XY, labels)
    
    # Silhouette plot
    y_ticks = []
    y_lower, y_upper = 0, 0
    for i, cluster in enumerate(np.unique(labels)):
        cluster_silhouette_vals = silhouette_vals[labels == cluster]
        cluster_silhouette_vals.sort()
        y_upper += len(cluster_silhouette_vals)
        ax1.barh(range(y_lower, y_upper), cluster_silhouette_vals, edgecolor='none', height=1)
        ax1.text(-0.03, (y_lower + y_upper) / 2, str(i + 1))
        y_lower += len(cluster_silhouette_vals)

    # Get the average silhouette score and plot it
    avg_score = np.mean(silhouette_vals)
    ax1.axvline(avg_score, linestyle='--', linewidth=2, color='green')
    ax1.set_yticks([])
    ax1.set_xlim([-0.1, 1])
    ax1.set_xlabel('Silhouette coefficient values')
    ax1.set_ylabel('Cluster labels')
    ax1.set_title('Silhouette plot for the various clusters', y=1.02);
    plt.suptitle('Silhouette analysis using k = '+str(k),fontsize=16, fontweight='semibold', y=1.05);
    
    # Scatter plot of data colored with labels

#     ax2.set_xlabel('Fear')
#     ax2.set_ylabel('Excitement')
#     ax2.set_zlabel('Happiness')
#     ax2.scatter(X_std[:, 0], X_std[:, 1], c=labels)
#     ax2.scatter(centroids[:, 0], centroids[:, 1], marker='*', c='r', s=250)
#     ax2.set_xlim([-2, 2])
#     ax2.set_xlim([-2, 2])
#     ax2.set_xlabel('Eruption time in mins')
#     ax2.set_ylabel('Waiting time to next eruption')
#     ax2.set_title('Visualization of clustered data', y=1.02)
#     ax2.set_aspect('equal')
    plt.tight_layout()
    
    print('k:', k, 'davies_bouldin_score:', davies_bouldin_score(XY, labels))
    davies.loc[i,'K'] = k
    davies.loc[i,'davies'] = davies_bouldin_score(XY, labels)
    print()
    
```

```python
from sklearn.mixture import GaussianMixture

# AIC, BIC lower is better !
n_components = np.arange(1, 12)
models = [GaussianMixture(n, covariance_type='full').fit(XY)
          for n in n_components]

plt.plot(n_components, [m.bic(XY) for m in models], label='BIC')
plt.plot(n_components, [m.aic(XY) for m in models], label='AIC')
plt.legend(loc='best')
plt.xlabel('n_components');
plt.show()
```

```python
aics = []
bics = []

for n in range(2, 10):
    gmm = GaussianMixture(n_components=n)
    labels = gmm.fit_predict(XY)
    
    colors = []
    for i in range(0, len(XY)):
        if labels[i] == 0:
            colors.append('r')
        elif labels[i] == 1:
            colors.append('b')
        elif labels[i] == 2:
            colors.append('g')
        elif labels[i] == 3:
            colors.append('orange')
        else:
            colors.append('pink')

    # davies_bouldin_score: lower is better
    print('k:', n, ', davies_bouldin_score:', davies_bouldin_score(XY, labels))
    
    plt.scatter(XY[:,0],XY[:,1],c=colors)
    plt.ylabel('Arousal')
    plt.xlabel('Valence')
    plt.title('K='+str(n))
    plt.savefig('kmean2.png')
    plt.show()
    
    aics.append(gmm.aic(XY))
    bics.append(gmm.bic(XY))
    

    
plt.plot(aics, label='aic')
plt.plot(bics, label='bic')
plt.legend()
plt.show()
```
